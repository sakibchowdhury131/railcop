```python
import numpy as np
np.random.seed(1)

import tensorflow as tf
tf.random.set_seed(1)
```

```python
import os
from tensorflow import keras
import matplotlib.pyplot as plt
import math
import pandas as pd
```

```python
label0 = pd.read_csv('/content/drive/MyDrive/RailCop/New Files/Sakib/data/train_7_label0.csv')
label1 = pd.read_csv('/content/drive/MyDrive/RailCop/New Files/Sakib/data/train_7_label1.csv')
print(label0)
```

```python
__label0 = label0.to_numpy()
__label1 = label1.to_numpy()[0:5300, :]
print(__label0)
```

```python
__label1.shape
```

```python
Y0 = np.zeros(__label0.shape[0])
Y1 = np.ones(__label1.shape[0])
print(Y0.shape)
print(Y1.shape)
```

```python
X = np.asarray(np.concatenate((__label0, __label1), axis = 0), dtype=np.float32)
Y = np.asarray(np.concatenate((Y0, Y1), axis = 0), dtype = np.float32)
print(X.shape)
print(Y.shape)
```

```python
from sklearn.utils import shuffle
X, Y = shuffle(X,Y)
```

```python
def network():
  __input = keras.Input((X.shape[1], 1))
  __conv1 = keras.layers.Conv1D(filters = 32, kernel_size = 3, activation = 'tanh')(__input)
  __dense2 = keras.layers.Dense(16, activation = 'tanh')(__conv1)
  __output = keras.layers.Dense(1, activation = 'sigmoid')(__dense2)

  model = keras.Model(inputs = __input, outputs = __output)
  return model 



model = network()
model.summary()
```

```python
model.compile(loss='binary_crossentropy',
                      optimizer= keras.optimizers.Adam(lr=0.0008),
                      metrics=['accuracy'])
```

```python
__modelVersion = 'MLP1.0'

weight_saver = tf.keras.callbacks.ModelCheckpoint('/content/drive/MyDrive/RailCop/New Files/Sakib/models/'+__modelVersion, monitor='val_accuracy', 
                                save_best_only=True, save_weights_only=False, mode= 'max')

annealer = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5,
                              patience=5, min_lr=0.00001,mode= 'min')


hist = model.fit( x = X , 
                  y = Y, 
                  batch_size = 64,
                  epochs = 600,
                  callbacks = [weight_saver , annealer],
                  validation_split = 0.2,
                  shuffle = True,
                  verbose = 1)
```
